# "Height" and "Weight" columns

```python
# changing the name of the "Height" and "Weight" column

df["Height"] = df["Height"].str.replace("cm","")
df.rename(columns={"Height":"Height Per cm"},inplace=True)
df["Weight"] = df["Weight"].str.replace("kg","")
df.rename(columns={"Weight":"Weight Per kg"},inplace=True)
```

```python
# Convert all values to be in cm and to be float

def convert_to_cm(height):
    if "'" in height and '"' in height:
        feet, inches = height.split("'")
        feet = int(feet)
        inches = int(inches.replace('"', ''))
        total_inches = feet * 12 + inches
        cm = total_inches * 2.54
        return cm
    else:
        return float(height) if height.replace('.', '', 1).isdigit() else height

# Apply the conversion function to the 'Height' column
df['Height Per cm'] = df['Height Per cm'].apply(convert_to_cm)
df["Height Per cm"] = df["Height Per cm"].astype(float)
```

```python
# Making sure that all values are ok

df["Height Per cm"].unique()
```

```python
# Convert all values to be in kg and to be float

def convert_to_kg(value):
    if isinstance(value, str) and "lbs" in value:
        value = float(value.replace("lbs", "")) * 0.45359237
        return value
    else:
        return float(value)
df["Weight Per kg"] = df['Weight Per kg'].apply(convert_to_kg)
```

```python
# Making sure that all values are ok

df["Weight Per kg"].unique()
```

# "Value", "Wage" and "Release Clause" columns

```python
# Removing the "€" sign to make the column easy to be processed

df["Value"] = df["Value"].str.replace("€","")
df["Wage"] = df["Wage"].str.replace("€","")
df["Release Clause"] = df["Release Clause"].str.replace("€","")
```

```python
# This cell and the 5 after it specialized in Dealing with "Value", "Wage" and "Release Clause" columns to be numerical

# Making 3 other columns to help me in my process

def create_new_column(value):
    if 'M' in value:
        return True
    elif 'K' in value:
        return False

df['Value_'] = df['Value'].apply(create_new_column)
df['Wage_'] = df['Wage'].apply(create_new_column)
df['Release Clause_'] = df['Release Clause'].apply(create_new_column)
```

```python
# Removing values of "M" and "K"

df["Value"] = df["Value"].str.replace("M","")
df["Value"] = df["Value"].str.replace("K","")
df["Wage"] = df["Wage"].str.replace("M","")
df["Wage"] = df["Wage"].str.replace("K","")
df["Release Clause"] = df["Release Clause"].str.replace("M","")
df["Release Clause"] = df["Release Clause"].str.replace("K","")
```

```python
df["Value"].unique()
```

```python
# Changing the type to float

df["Value"] = df["Value"].astype(float)
df["Wage"] = df["Wage"].astype(float)
df["Release Clause"] = df["Release Clause"].astype(float)
```

```python
# Returning the values as it were but in numerical image

df['Value'] = df.apply(lambda row: row['Value'] * (1000000 if row['Value_'] else 1000) if row['Value_'] is not None else row['Value'], axis=1)
df['Release Clause'] = df.apply(lambda row: row['Release Clause'] * (1000000 if row['Release Clause_'] else 1000) if row['Release Clause_'] is not None else row['Release Clause'], axis=1)
df['Wage'] = df.apply(lambda row: row['Wage'] * (1000000 if row['Wage_'] else 1000) if row['Wage_'] is not None else row['Wage'], axis=1)
```

```python
# Removing the useless columns after we did what we needed (Sorry i am not selfish but it is what it is)

df = df.drop(columns="Value_")
df = df.drop(columns="Wage_")
df = df.drop(columns="Release Clause_")
```
